# N dimensional surface

An $N$ dimensional sphere is given, and we need to distribute $M$ points on its surface and return an array of the vectors of descartes coordinates.

```python
import matplotlib.pyplot as plt
import math
import numpy
from timeit import default_timer
a = 20
N = 30000
alpha = 100/N
```

## Method 1

Create points in an ellipsoid on an ellipse directly. The length (radius) is calculated using the ellipse parameter.

```python
numpy.random.seed(0)
x = numpy.random.normal(size=N, scale=a)
y = numpy.random.normal(size=N)

length = numpy.sqrt(numpy.square(x/a) + numpy.square(y))
x /=length
y /=length

plt.figure(figsize=(30,30/a+1))
plt.plot(x, y,"ro",alpha=alpha)
plt.gca().set_aspect(aspect=1)
plt.suptitle("Method 1")
plt.show()
```

On a circle, this would look like:

```python
x = numpy.random.normal(size=N, scale=a)
y = numpy.random.normal(size=N)

length = numpy.sqrt(numpy.square(x) + numpy.square(y))
x /=length
y /=length

plt.figure(figsize=(5,5))
plt.plot(x, y,"ro",alpha=alpha)
plt.gca().set_aspect(aspect=1)
plt.suptitle("Method 1 on a circle")
plt.show()
```

## Method 2
Create points on a circle, then compress the circle.

```python
numpy.random.seed(0)
x = numpy.random.normal(size=N)
y = numpy.random.normal(size=N)

length = numpy.sqrt(numpy.square(x) + numpy.square(y))
x /=length/a
y /=length

plt.figure(figsize=(30,30/a+1))
plt.plot(x, y,"ro",alpha=alpha)
plt.gca().set_aspect(aspect=1)
plt.suptitle("Method 2")
plt.show()
```

It can be seen that the two methods produces the same result.

# Method 3
Create points on a reversly elongated circle, then compress the circle.

```python
numpy.random.seed(0)
print("1/a**0.5 = ",1/a**0.5)
x = numpy.random.normal(size=N,scale=1 - 0.5*(a-1) + 0.375*(a-1)**2)
y = numpy.random.normal(size=N)

length = numpy.sqrt(numpy.square(x) + numpy.square(y))
x /=length/a
y /=length

plt.figure(figsize=(30,30/a+1))
plt.plot(x, y,"ro",alpha=alpha)
plt.gca().set_aspect(aspect=1)
plt.suptitle("Method 3")
plt.show()
```

On the circle, this would look like:

```python
numpy.random.seed(0)
x = numpy.random.normal(size=N,scale=1/a**0.5)
y = numpy.random.normal(size=N)

length = numpy.sqrt(numpy.square(x) + numpy.square(y))
x /=length
y /=length

plt.figure(figsize=(5,5))
plt.plot(x, y,"ro",alpha=alpha)
plt.gca().set_aspect(aspect=1)
plt.suptitle("Method 3 on a circle")
plt.show()
```
